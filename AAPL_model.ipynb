```python
import requests
import pandas as pd

def get_data(ticker_id, api_key):
    url = f"https://api.tiingo.com/tiingo/daily/{ticker_id}/prices"
    params = {
        'startDate': '2010-01-01',
        'format': 'json',  # Use 'json' instead of 'csv' for JSON response
        'resampleFreq': 'daily',
        'token': api_key,
    }
    response = requests.get(url, params=params)
    data = response.json()
    return data

    
api_key = "5f9b868e8158bbcbed75f0549ff6f14b7a62b8f8"
ticker_symbol = "MSFT"

data = get_data(ticker_symbol, api_key)
data1 = pd.DataFrame(data)
data1
```
